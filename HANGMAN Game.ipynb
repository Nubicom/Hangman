```python
import turtle as t
t.reset()
t.speed(1)
import random
import time

# Initial Steps to invite in the game:
print("COMPUTER SCIENCE ACADEMY AFRICA 2022 WORKSHOP PROJECT\nHANGMAN GAME by GROUP 21\nCHIDINMA NWALIEZE\nDAMILARE OBAFEMI\nHALIMAT AMZAT\nLUKMON OTUNUBI\nOLAYINKA OYINLOYE\n")
name = input("Enter your name: ")
print(f"{name}! Welcome to the HANGMAN game!\nBest of Luck!\n")
print("The game is about to begin!\n")
time.sleep(2)


# The parameters we require to execute the game:
with open ("HANGMAN word.txt") as f:
    guesslist=[]
    for line in f:
        l = line.strip()
        guesslist.append(l)
# print(guesslist)
    
word= random.choice(guesslist)
length = len(word)
count = 0
display = '-' * length
already_guessed = []
# print(word)

num = "1234567890"

# Initializing all the conditions required for the game:
def hangman():
    global count
    global display
    global word
    global already_guessed
    global num 
    limit = 5
    guess = input(f"The Hangman Word is : {display} Enter your guess: \n")
    guess = guess.strip()
    guess = guess.lower()
    if len(guess) == 0 or len(guess) >= 2 or guess in num:
        print("Invalid Input, Try a letter\n")
        hangman()


    elif guess in word:
        already_guessed.extend([guess])
        position = word.find(guess)
        word = word[:position] + "-" + word[position + 1:]
        display = display[:position] + guess + display[position + 1:]
        print(display + "\n")

    elif guess in already_guessed:
        print("Try another letter.\n")

    else:
        count += 1

        if count == 1:
            time.sleep(1)
            t.penup()
            t.goto(-100,-200)
            t.pendown()
            t.pensize(5)
            t.forward(200)
            t.backward(100)

            t.left(90) 

            t.forward(450)
            t.right(90) 
            t.forward(200)
            t.right(90)
            t.forward(25)
            t.right(90) #first miss
            print("Wrong guess. " + str(limit - count) + " guesses remaining\n")

        elif count == 2:
            time.sleep(1)
            def eye(col, rad):
                t.down()
                t.fillcolor(col)
                t.begin_fill()
                t.circle(rad)
                t.end_fill()
                t.up()


            # draw face
            t.fillcolor('yellow')
            t.begin_fill()
            t.circle(50)
            t.end_fill()
            t.up()
            t.left(90)

            # draw eyes
            t.goto(205.5, 192.5)
            eye('white', 8.5)
            t.goto(205.5, 192.5)
            eye('black', 2.5)
            t.goto(175.5, 192.5)
            eye('white', 8.5)
            t.goto(175.5, 192.5)
            eye('black', 2.5)

            # # draw nose
            t.goto(195.25, 170.5)
            eye('black', 4)

            # # # draw mouth
            t.goto(180.5, 170.5)
            # t.right(90)
            t.penup()
            t.forward(20)
            t.down()
            t.left(90)
            t.forward(40)
            t.backward(33.75)
            t.right(90) 
            t.up()

            # #draw tongue

            t.fillcolor('red')
            t.begin_fill()
            t.circle(10, 180)
            t.end_fill()
            t.penup()
            t.goto(197.25, 122.5)
            t.right(180)
            t.pendown() #second miss
            print("Wrong guess. " + str(limit - count) + " guesses remaining\n")

        elif count == 3:
            time.sleep(1)
            t.pensize(15)
            t.forward(25)
            t.right(60) 
            t.forward(50)
            t.backward(50)
            t.right(60) 
            t.backward(50)
            t.forward(50)
            t.left(120) #third miss
            print("Wrong guess. " + str(limit - count) + " guesses remaining\n")

        elif count == 4:
            time.sleep(1)
            t.forward(100)#fourth miss
            print("Wrong guess. " + str(limit - count) + " last guess remaining\n")

        elif count == 5:
            time.sleep(1)
            t.right(60) 
            t.forward(50)
            t.backward(50)
            t.right(60)
            t.backward(50)
            t.hideturtle()#fifth miss
            t.penup()
            t.goto(-100, -250)
            t.pendown()
            t.write("AWW! You got HANGED!",font=("Arial", 16, "bold"))
            

            print("Hahahahaha!!!!You got hanged!!!\n")
            print("The word was:",already_guessed,word)
    

    if word == '-' * length:
        #using graphics to draw a star
        t.reset()
        t.pensize(5)
        t.fillcolor("yellow")
        t.begin_fill()

        t.forward(50)
        t.left(60)
        t.forward(50)
        t.right(120)
        t.forward(50)
        t.left(60)
        t.forward(50)
        t.right(150)
        t.forward(50)
        t.left(90)
        t.forward(80)
        t.right(150)
        t.forward(80)
        t.left(60)
        t.forward(80)
        t.right(150)
        t.forward(80)
        t.left(90)
        t.forward(50)
        t.end_fill()
        t.hideturtle()
        t.penup()
        t.goto(-100, 100)
        t.pendown()
        t.write("BRAVO! You are a STAR",font=("Arial", 25, "bold"))
        t.hideturtle()
        print("Congrats! You have guessed the word correctly!\n You are a STAR")


    elif count != limit:
        hangman()

hangman()
```
